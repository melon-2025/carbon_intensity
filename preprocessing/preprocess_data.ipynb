# Imports

```python
import pandas as pd
import os
```

## Load Carbon Intensity datasets

```python
# base path
base_path = "../data/raw/"
years = [2021, 2022, 2023]  

# Read
df_CI_hourly = pd.concat([pd.read_csv(f"{base_path}CH_{year}_hourly.csv") for year in years], ignore_index=True)
```

### relative path to be written to an excel file

```python
relative_path = os.path.join("..", "data", "preprocessed", "processed_data.xlsx")
```

## Load energy consumption and production dataset

```python
path="../data/raw/"

df_energy = pd.concat([pd.read_excel(f"{path}EnergieUebersichtCH-{year}.xlsx",sheet_name='Zeitreihen0h15') for year in years], ignore_index=True)
```

```python
# carbon intensity
df_CI_hourly.head()
```

convert to date time, drop first row and reset index

```python
#Rename column
df_energy.rename(columns={'Summe endverbrauchte Energie Regelblock Schweiz\nTotal energy consumed by end users in the Swiss controlblock': 'End user energy consumtion(KWH)', 
                   'Summe produzierte Energie Regelblock Schweiz\nTotal energy production Swiss controlblock': 'Total energy production Swiss controlblock(KWH)',
                    'Summe verbrauchte Energie Regelblock Schweiz\nTotal energy consumption Swiss controlblock':'Total energy consumption Swiss controlblock(KWH)',
                    'Vertikale Einspeisung ins Übertragungsnetz Schweiz\nGrid feed-in Swiss transmission grid':'Grid feed-in Swiss transmission grid(KWH)',
                    'Netto Ausspeisung aus dem Übertragungsnetz Schweiz\nNet outflow of the Swiss transmission grid':'Net outflow of the Swiss transmission grid(KWH)'}, inplace=True)
```

```python
# rename
df_CI_hourly.rename(columns={'Datetime (UTC)':'Date'},inplace=True)
df_CI_hourly['Date']=pd.to_datetime(df_CI_hourly['Date'],dayfirst=True)
#set index
df_CI_hourly.set_index('Date',inplace=True)
```

```python
df_CI_hourly.head()
```

rename and drop unnecesary row

```python
df_energy.rename(columns={'Unnamed: 0':'Date'},inplace=True)
df_energy=df_energy.drop(index=0)
df_energy.reset_index(drop=True,inplace=True)
df_energy['Date']=pd.to_datetime(df_energy['Date'],errors='coerce',dayfirst=True)
df_energy.set_index('Date',inplace=True)
```

```python
df_energy.head()
```

resample 15min data to hourly data


```python

# resample the columns needed and aggregate them
df_energy_hourly = df_energy.resample('h').agg({
    'End user energy consumtion(KWH)': 'sum',  
    'Total energy production Swiss controlblock(KWH)' :'sum',
    'Total energy consumption Swiss controlblock(KWH)': 'sum',
    'Grid feed-in Swiss transmission grid(KWH)':'sum',
    'Net outflow of the Swiss transmission grid(KWH)': 'sum'  
})
```

keep the CI to merge

```python
df_CI_tomerge=df_CI_hourly[['Carbon Intensity gCO₂eq/kWh (LCA)','Carbon Intensity gCO₂eq/kWh (direct)','Low Carbon Percentage','Renewable Percentage']]
```

## merge the two datasets

```python
df=pd.merge(df_CI_tomerge,df_energy_hourly,on='Date')
```

Write to an excel file

```python
df.to_excel(relative_path, index=True)
```
